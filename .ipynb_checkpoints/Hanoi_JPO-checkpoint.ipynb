# Le jeu des Tours de Hanoï

D'après [wikipedia](https://fr.wikipedia.org/wiki/Tours_de_Hano%C3%AF) :

Les tours de Hanoï (originellement, la tour d'Hanoïa) sont un jeu de réflexion imaginé par le mathématicien français Édouard Lucas, et consistant à déplacer des disques de diamètres différents d'une tour de « départ » à une tour d'« arrivée » en passant par une tour « intermédiaire », et ceci en un minimum de coups, tout en respectant les règles suivantes :

- on ne peut déplacer plus d'un disque à la fois ;
- on ne peut placer un disque que sur un autre disque plus grand que lui ou sur un emplacement vide.

On suppose que cette dernière règle est également respectée dans la configuration de départ.

### Aide pour comprendre l'algorithme

Il existe plusieurs algorithmes (itératifs et récursif) pour résoudre ce problème. Si leurs implémentations sont simples, leurs fonctionnements peuvent être durs à appréhender.

<details>
    <summary><b>Clique ici</b> si tu veux de l'aide ;)</summary>

  ---
  On va voir une approche itérative simple qui consiste à réaliser des déplacements en boucle entre toutes les paires de poteaux jusqu'à que la partie soit terminée. Les paires de poteaux sur lesquelles itérer jusqu'à la fin du jeu sont les suivantes :
    
  - (A, B)
  - (A, C)
  - (B, C)
    
  > Lorsque l'on doit réaliser un déplacement pour le couple (A, B), par exemple, il se peut que le mouvement du poteau A vers le poteau B soit impossible. Il faut alors jouer B vers A.
    
  Par exemple, s'il y a 3 disques en jeu, on va boucler sur les couples (A, C), (A, B) et (B, C). Voici les coups qu'il faut jouer :
  
  - A -> B
  - A -> C
  - B -> C
  
  On a joué toutes les paires, mais la partie n'est pas terminé. On recommence :
  
  - A -> B
  - A -> C (ici, le mouvement A -> C est impossible : on joue donc C -> A)
  - B -> C (ici, le mouvement B -> C est impossible : on joue donc C -> B)
  
  On a joué toutes les paires, mais la partie n'est pas terminé. On recommence :
    
  - A -> B
    
  La partie s'arrête alors car l'ensemble des disques a été déplacé.
    
  **N'hésites pas à appeler la fonction ``is_move_valid`` implémentée précédemment !**
</details>
